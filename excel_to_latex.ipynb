```python
import os
import regex
import pandas

folder_path = 'chapters/DataPercobaan'
# Cek jika direktori ada, jika tidak, buat direktori tersebut
if not os.path.exists(folder_path):
    os.makedirs(folder_path)
    print(f"Direktori '{folder_path}' telah dibuat.")
    # Anda mungkin perlu menambahkan file Excel contoh di sini
    # Contoh: pd.DataFrame({'Waktu (detik)': [0.1], 'Tinggi (cm)': [10], 'Sensor (cm)': [35]}).to_excel(os.path.join(folder_path, 'contoh_file.xlsx'), index=False)

xlsx_files = [f for f in os.listdir(folder_path) if f.endswith('.xlsx')]
print(f"Found {len(xlsx_files)} Excel files in '{folder_path}'.")

# create caption
def create_caption(file_name):
    caption =  file_name.replace('.xlsx', '').replace('_', ' ').title()
    return caption

latex_tables = """"""
for file in xlsx_files:
    caption = create_caption(file)
    print(f"Processing file: {file} with caption: {caption}")
    file_path = os.path.join(folder_path, file)
    df = pd.read_excel(file_path)
    # remove last column
    df = df.iloc[:, :-1] if df.shape[1] > 0 else df

    latex_str = df.to_latex(
        index=False,
        escape=True,
        header=True,
        na_rep='',
        bold_rows=False,
        column_format='|' + '|'.join(['c'] * len(df.columns)) + '|',
        longtable=True,
        multicolumn=False,
        multicolumn_format='c',
        multirow=False,
        caption=caption,
        label=None,
        position='htbp',
        float_format=None,
        decimal='.'
    )

    # hapus toprule,midrule dan bottomrule
    latex_str = latex_str.replace('\\toprule\n', '').replace('\\midrule\n', '').replace('\\bottomrule\n', '')

    # --- KODE BARU UNTUK FORMATTING (VERSI PERBAIKAN) ---
    # 1. Menambahkan \hline pada baris baru setelah baris \caption
    # Menggunakan \1 untuk merujuk pada grup pertama yang ditangkap (seluruh baris caption)
    latex_str = regex.sub(r'(^\\caption.* \\\\$)', r'\1\n\\hline', latex_str, flags=regex.MULTILINE)

    # 2. Menambahkan spasi dan \hline di akhir setiap baris data dan header
    # Menggunakan \1 untuk merujuk pada grup pertama yang ditangkap (seluruh baris data/header)
    latex_str = regex.sub(r'(^.*&.* \\\\$)', r'\1 \\hline', latex_str, flags=regex.MULTILINE)
    # --------------------------------------------------------

    print(f"Latex for {file}:\n{latex_str}\n")
    # add to latex_tables
    latex_tables += f"""{latex_str}"""

    # save to file
    latex_filename = file.replace('.xlsx', '.tex')
    latex_path = os.path.join(folder_path, latex_filename)
    with open(latex_path, 'w', encoding='utf-8') as f:
        f.write(latex_str)
```
